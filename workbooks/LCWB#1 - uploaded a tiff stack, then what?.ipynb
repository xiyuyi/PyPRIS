# LC workbook \#1 
You've upload a tiff stack and the psf stack for your dataset, you would like to perform PRIS fitting on all the frames of your tiff stack, using the uploaded psf.tif as the PSF model. How to do it step by step? Here is a workbook for it.

### steps:
[step1](prep) -- [step2](decomp) -- [step3](optimize) -- [step4](submit_para_sweep)
    1. Upload the tiff stack, and the psf.tif, and prepare.    
    2. Decompose the tiff stack into many single tiff images (easy for integration in to the analysis pipeline)
    3. Optimize the pris fitting parameters using a single tiff image [small scale parameter sweeping]
    4. Submit -debug jobs on LC to perform pris analysis for the collection of tickets for one sweep.
    5. Check and visualize the parameter sweeping results
    6. Repeat steps 3-5 until a desired set of fitting parameters are found.
        -- Ensure fitting quality and reliability
            - The sesidual follows gaussian distribution
            - No dummy emitters in the background identified by the algorithm to patch for background component.
            - make sure the iteration do not blow up with calculate with 1.5 to 2 fold of iteration steps.
        -- Ensure convergence across all pris iterations
            - residual needs to converge.
            - background component needs to converge.
    7. Choose another frame, and test with the current ticket, fit and visualize just to confirm. 
    8. Prepare the full packet of pris ticket with the chosen parameters.
    9. Submit all jobs
    10. Use the task monitor to make sure all the calculations are finished
    11. convert all fitting results to location coordinates, compile data on LC.
    12. download and visualize data

### Step 2. Decompose the tiff stack into many single tiff files
<a id='decomp'></a>

```python
# Define the paths and the file names
dpath='/p/lscratchh/yi10/spt-ras/simu3-benchmark-data'
batchN=15

# Define the full path for the tiff stack:
tiff_stack_file_name = dpath + '/chara_Batch'+str(batchN)+'_noisy_bg160_timeBin1.tif'

# Define the folder to store the single tiff images decomposed from the tiff stack
tiff_stack_otpt_path = dpath + '/prep_for_PRIS/Pack_chara_Batch'+str(batchN)+'_noisy_bg160_timeBin1'

# Define the folder to store all the tickets
tiff_stack_tckt_path = dpath + '/prep_for_PRIS/tickets_chara_Batch'+str(batchN)+'_noisy_bg160_timeBin1'

# now make the folders in case they do not exist
try:
    os.mkdir(dpath+'/prep_for_PRIS')
except:
    pass

try:
    os.mkdir(tiff_stack_otpt_path)
except:
    print('making individual tiff images pack folder failed')
    pass

try:
    os.mkdir(tiff_stack_tckt_path)
except:
    print('making ticket pack folder failed')
    pass


# Now decompose the tiff stack into individual tiff images, store with names as frame1.tif, frame2.tif, ...
stack=io.imread(tiff_stack_file_name);
for i in np.arange(len(stack)):
    io.imsave(tiff_stack_otpt_path+'/frame'+str(i)+'.tif', stack[i])
```

### Step 3. Optimize the PRIS fitting parameters (ticket configuration) 
<a id='optimize'></a>
Before we perform PRIS fitting of all the images from the entire tiff stack, we will first try to optimize the PRIS fitting parameters. These fitting parameters are configured and packaged as a ticket object to be taken in by the pris algorithm. Different classes of tickets are predefined, and here we are working with a type of pris tickets that deals with single plane observations.

#### Step 3.1. Prepare basic parameters for the ticket.


```python
# get a ticket of 'SinglePlaneTicket' class.
ticket = SinglePlaneTicket()
frameN = 0;
# define basic properties of the ticket that we do not want to explore in this parameter optimization process.
ticket.datapath = tiff_stack_otpt_path # this the folder where the data is stored.
ticket.observation_channel_n = 'SinglePlane'
ticket.plane1_path = "{}/frame{}.tif".format(ticket.datapath,str(frameN))
ticket.psf_path = "{}/psf.tif".format(ticket.datapath)
ticket.psf_norm_factor = 1 # The PSF normalization factor is a number that 
                              # normalizes the total energy of a PSF to be 1
                              # But experimental calibration data, the PSF matrix is often noisy, 
                              # so an estimate would be enough.
            # when the background is too strong, we change this to a smaller number.. forces the scale of the signal measures to be higher

# specification of the psf stack
ticket.psfz0 = 101  # the count of the center plane, define it as z=0 in the psf coordinates system.
ticket.observer_edge_padding = False # switch to True when PSF has high bg component.

# configure the initial candidate pool of this pris ticket
ticket.init_candidates_intervals = list([1, 1, 1])

ticket.init_ax0_range = list([0, 1])
ticket.init_ax1_range = list([0, 128])
ticket.init_ax2_range = list([0, 128])

# debug configurations"
ticket.observer_debugger = False
ticket.tobserver_edge_padding = True

# output settings"
ticket.ticket_pack_folder = tiff_stack_tckt_path

# linbreg configurations"
ticket.linbreg_alpha = LinBreg("X")
ticket.linbreg_alpha.debug = True
ticket.linbreg_alpha.deep_debug = False

# other parameters
ticket.PRIS_iter_end = 5
ticket.expansion = False

# initial configuration for linbreg object.
ticket.linbreg_alpha.maxit = 3000
ticket.linbreg_alpha.maxit_delta = 1000 # additional threshold to the maxit after each pris iteration.
ticket.linbreg_alpha.it_check_rem = 0
ticket.linbreg_alpha.debug_it_int = 500
ticket.linbreg_alpha.kick.ints = ticket.linbreg_alpha.maxit # dont kick when density is high.
ticket.linbreg_alpha.kick.eval_ints = 2000
ticket.linbreg_alpha.kick.flag = True
ticket.linbreg_alpha.kick.thres = 0.1
ticket.linbreg_alpha.save_obj_int = 2000
ticket.linbreg_alpha.flag_bg_allow_negative = False # force background to not be negative
ticket.linbreg_alpha.e = True
ticket.linbreg_alpha.PyPRIS_iter = 0
ticket.linbreg_alpha.path_0 = '.'
ticket.linbreg_alpha.stopping_loghistpercdelres_thres = -14
ticket.linbreg_alpha.mu = 1e6
ticket.linbreg_alpha.alpha = 1e-6
ticket.linbreg_alpha.auto_mu = True
ticket.linbreg_alpha.auto_bg = True
ticket.bg_scaling_coef = 10
```

#### Step 3.2. Create ticket packet for parameter optimization
loop over different options of auto bg fold, and auto mu fold. and create the ticket packet.

```python
# define a list of mu-fold and bg-folds to sweep through:
muFolds=[50, 100, 200]
bgFolds=[1, 1.2, 2, 4]


# now calculate the packet of tickets for parameter optimization.
for auto_mu_fold in muFolds:
    for auto_bg_fold in bgFolds:
        ticket_new = copy.deepcopy(ticket)
        ticket_new.name="frame"+str(frameN)+'_muFold'+str(auto_mu_fold)+'_bgFold'+str(auto_bg_fold)
        ticket_new.linbreg_alpha.PyPRIS_name = ticket_new.name
        ticket_new.linbreg_alpha.auto_mu_fold = auto_mu_fold;
        ticket_new.linbreg_alpha.auto_bg_fold = auto_bg_fold  # increase number to make the bg compoentn more competitive than random background 'non-specific-emitters'.
                                                    # use a larger number when you think the background component in the calibrated PSF matrix
            # is significantly smaller than tht in the data. 
            # you may consider using a number smaller than 1 when you thin the background component in PSF calibration is big or comparable.
        try:
            if not os.path.exists("{}/{}".format(ticket_new.ticket_pack_folder, ticket_new.name)):
                os.mkdir("{}/{}".format(ticket_new.ticket_pack_folder, ticket_new.name))
        except OSError:
            pass
        with open("{}/{}/Go.pris_ticket".format(ticket_new.ticket_pack_folder, ticket_new.name), "wb") as f:
            pickle.dump(ticket_new, f, pickle.HIGHEST_PROTOCOL)
```
